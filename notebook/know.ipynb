## What is RAG ?


Rag is the process of optimizing the output of a large language model, so it references an authoritative knowledge base outside of its training data sources before generating a response.  LLMS are trained on vast volumes of data and use billions of parameters to generate original output for tasks like answering questions, translating languages and completing sentences. RAG extends the already powerful capablilites of LLMS to specific domanins or an organization's internal knowledge base, all without the need to retrain the model. It is a cost-effective approach to improving LLM output so it remains relevant, accurate and useful in various contexts. 

RAG is the solution to LLM Hallucination, instead of trying to finetune billions of parameter with new data. What do we do?
